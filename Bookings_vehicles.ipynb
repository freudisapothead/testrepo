```python
data = pd.read_csv("C:/Users/gokhan.elbistan.ABC01LP014/Desktop/DFDS Booking/Booking import/companies/aractip.csv", header=0, sep=';')
df1 = pd.DataFrame(data)
```

```python
df1.rename(columns={'type_id': 'vehicle_type_id'},  inplace=True)
```

```python
df1['vehicleid'].replace('NaN', None, inplace=True)
```

```python
def generate_update_statements(dataframe):
  """
  Generates UPDATE statements for each row in a Pandas DataFrame,
  to update the vehicle_type_id in the vehicles_vehicle table.

  Args:
    dataframe: A Pandas DataFrame.

  Returns:
    A list of strings, each containing an UPDATE statement.
  """
  statements = []
  for index, row in dataframe.iterrows():
    if row['vehicle_id'] != 'NaN' and row['vehicle_id'] is not None:
      statement = f"UPDATE public.vehicles_vehicle SET vehicle_type_id = '{row['vehicle_type_id']}' WHERE id = '{row['vehicle_id']}';"
      statements.append(statement)
  return statements

# Generate update statements for df1
update_statements = generate_update_statements(df1)

# Print the statements
for statement in update_statements:
  print(statement)

# Alternatively, you can write the statements to a file
with open("update_statements_for_vehicles.sql", "w") as file:
  file.writelines(update_statements)
```
