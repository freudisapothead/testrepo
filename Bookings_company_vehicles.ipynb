```python
df1['vehicleid'].replace('NaN', None, inplace=True)
```

```python
import uuid

# Generate a UUID for each row
df1.insert(0, 'id', [uuid.uuid4() for _ in range(len(df1))])

# Print the updated dataframe
df1
```

```python
import datetime

# Get the current datetime
current_datetime = datetime.datetime.now()

# Add the created_date and updated_date columns after the id column
df1.insert(1, 'created_date', current_datetime)
df1.insert(2, 'updated_date', current_datetime)

# Print the updated dataframe
df1
```

```python
df1.rename(columns={'firmaid': 'company_id', 'vehicleid': 'vehicle_id'},  inplace=True)
```

```python
current_order = df1.columns
desired_order = ['id', 'created_date', 'updated_date', 'company_id', 'vehicle_id']
df1 = df1[desired_order]
```

```python
df1
```

```python
dataframe = df1
```

```python
import os

def generate_insert_statements(dataframe):
  """
  Generates INSERT statements for each row in a Pandas DataFrame,
  including quotes for all values.

  Args:
    dataframe: A Pandas DataFrame.

  Returns:
    A list of strings, each containing an INSERT statement.
  """
  columns = dataframe.columns
  statements = []
  for index, row in dataframe.iterrows():
    values = [f"'{value}'" for value in row]
    statement = f"""
INSERT INTO public.vehicles_vehiclecompany(
  {", ".join(columns)}
) VALUES (
  {", ".join(values)}
);
"""
    statements.append(statement)
  return statements

# Generate insert statements for filtered_combined_df
insert_statements = generate_insert_statements(df1)

# Print the statements
for statement in insert_statements:
  print(statement)

# Write the statements to a file
file_path = "insert_statements_for_company_vehicles.sql"
with open(file_path, "w") as file:
  file.writelines(insert_statements)

# Check if the file was created
if os.path.exists(file_path):
  print("File created successfully.")
else:
  print("File creation failed.")
```
